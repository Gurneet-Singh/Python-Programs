Your high school is filling out class rosters and is creating two sections of health class. Append, merge, sort, and split DataFrames to create the final class rosters!<br> 
Start by adding the provided information into the DataFrame. <br>
Syed registered late, so don't forget to append his information as well.

```
import pandas as pd
students1 = {"Name":['Alberto', 'Gino', 'Ryan', 'Eesha'], 
             "Grade":[10, 11, 12, 10],
             "GPA":[4.0, 3.4, 2.8, 3.7]}
syed = {"Name":"Syed", "Grade":11, "GPA":3.2}
```

```
students1['Name'].append('Syed')
students1['Grade'].append(11)
students1['GPA'].append(3.2)
print(students1)
```

Now add information for the next 5 students enrolling in health to the DataFrame.

```
name_2 = ['Andi','Brielle','Tom', 'Selma', 'Ron']
for i in name_2:
  students1['Name'].append(i)

grade_2 = [12,12,11,10,11]
for i in grade_2:
  students1['Grade'].append(i)

gpa_2 = [2.9, 3.8, 3.6, 3.9, 2.7]
for i in gpa_2:
  students1['GPA'].append(i)
```

```
print(students1)
```

```
df = pd.DataFrame(students1)
display(df)
```

Now let's add 10 new students who will be taking health to the DataFrame. <br>
These students are Freshmen so their grade will be set to 9 and GPA to 0.

```
name_3 = ['Sam','Jenny','Tana','Steve','Carlos','Akash','Cher','Shilpa','Brad','Quinn']
grade3 = [9 for i in range(len(name_3))]
gpa3 = [0.0 for i in range(len(name_3))]

for i in name_3:
  students1["Name"].append(i)

for i in grade3:
  students1["Grade"].append(i)

for i in gpa3:
  students1["GPA"].append(i)
```

```
print(students1)
```

```
df = pd.DataFrame(students1)
display(df)
```

Now generate three digit ID numbers for each student as a new column in the DataFrame.

```
import random
df["ID"] = 1
for index, row in df.iterrows():
  df.loc[index, "ID"] = random.randint(100, 999)
display(df)
```

Randomly split the class into two sections and create DataFrames `period1` and `period2`. Sort each DataFrame according to `Grade` and reset the indices.

Hint: Sample from the original DataFrame to create `period1`. Then remove (drop) `period1` indices from the original DataFrame to get `period2`. You can get the indices of a DataFrame by calling `df.index`

```
per1 = df.sample(int(len(df)/2))
per1 = per1.sort_values(by=["Grade"], ascending=True).reset_index(drop=True)
per2 = df.drop(per1.index)
per2 = per2.sort_values(by=["Grade"], ascending=True).reset_index(drop=True)
```
