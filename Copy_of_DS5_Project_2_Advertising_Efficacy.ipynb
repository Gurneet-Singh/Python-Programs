Pearson's Coefficient measures the strength of a linear correlation between two variables. It ranges between -1 and 1. -1 represents a perfect negative correlation. 1 represents a perfect positive correlation. 0 means there's no correlation at all. 

Using the `strength` function below, quantify the strength of each correlation. Do they seem to match your findings in the graphs?

Student Analysis:


```python
# Starter Code
from scipy import stats
def strength(variable1, variable2):
  return stats.pearsonr(variable1,variable2)[0]
```

```python
strength(df["TV"], df["Sales"])
```

```python
strength(df["Radio"], df["Sales"])
```

```python
strength(df["Newspaper"], df["Sales"])
```

Finally, estimate the slope of the best fit line for the type of advertising that has the strongest correlation. This will tell us the amount of money we make for each dollar that we spend on this type of advertising. Estimate two coordinates on the graph `(x1, y1)` and `(x2, y2)`. Then compute the slope by calculating `(y2-y1)/(x2-x1)`.

In order to make an accurate prediction consider these tips:<br>
`truncate = False` to avoid cutting off the line of best fit<br>
`plt.xlim(min, max)` to set the scale for the x axis <br>
`plt.grid(True)` to add gridlines <br>

Student Analysis:

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(df['TV'],df['Sales'])

ax = sns.regplot(x="TV", y="Sales", data=df, color='b', 
 line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope,intercept)}) #line_kws={'label':'y='+str(slope)+'x'+str(intercept)}
 
ax.legend()
```
