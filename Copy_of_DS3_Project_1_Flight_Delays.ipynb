```
delays = e.copy()

for column in delays.columns:
  if "Delay" not in column:
    delays.drop(column, inplace = True, axis = 1)

display(delays)
delays.describe()
```

Now let's investigate the problem further. Let's determine the number of flights in our dataset that occurred on each day of the month. Group the original DataFrame by `DayofMonth` and use `count()` to get a count of the number of flights for each day. Determine which days had the minimum and maximum number of flights. 

```
#display(e.groupby('DayofMonth').count())
```

```
#display(e)
grouped = e.groupby("DayofMonth").count()["Year"]
maxcount = grouped.max()
maxindx = grouped.idxmax()
```

```
grouped = e.groupby("DayofMonth").count()["Year"]
mincount = grouped.min()
minindx = grouped.idxmin()
print(mincount, minindx)
```

The Department of Transportation would like the know how the number of flights impacts delays. Make two new DataFrames for flights during the day of the month with the highest and lowest number of flights. Then output the mean `ArrDelay` for each DataFrame.

```
e[e["DayofMonth"] == maxindx]["ArrDelay"].mean()
```

```
e[e["DayofMonth"] == minindx]["ArrDelay"].mean()
```
